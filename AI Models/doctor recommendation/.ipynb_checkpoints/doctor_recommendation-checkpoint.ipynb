```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import pickle

# Load the Excel file
file_path = "Specialist.xlsx"  
df = pd.read_excel(file_path)

# Drop the first row if needed
df = df.drop(index=0)

# Encode the 'Disease' column to numeric values
label_encoder = LabelEncoder()
df['Doctor'] = label_encoder.fit_transform(df['Doctor'])

# Separate X (features) and Y (target)
X = df.drop('Doctor', axis=1)  # All columns except 'Disease'
Y = df['Doctor']  # The 'Disease' column as target

best_accuracy = 0
best_model = None
best_Y_test = None
best_Y_pred = None

# Run the training 10 times
for i in range(10):
    # Split the data into train and test sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=i)

    # Create a KNN classifier
    knn = KNeighborsClassifier(n_neighbors=5)

    # Train the model
    knn.fit(X_train, Y_train)

    # Predict using the test set
    Y_pred = knn.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"Iteration {i+1}: Model Accuracy = {accuracy * 100:.2f}%")

    # Check if this is the best model
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = knn
        best_Y_test = Y_test
        best_Y_pred = Y_pred

# Save the best model
with open('best_knn_model.pkl', 'wb') as f:
    pickle.dump(best_model, f)

# Print the test results and actual values of the best model
print("\nBest Model Test Results:")
print(f"Accuracy: {best_accuracy * 100:.2f}%")
print("Actual Values:", best_Y_test.values)
print("Predicted Values:", best_Y_pred)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, mean_squared_error
import pickle
import numpy as np

# Load the Excel file
file_path = "Specialist.xlsx"  # Replace with your actual file path
df = pd.read_excel(file_path)

# Drop the first row if needed
df = df.drop(index=0)

# Encode the 'Disease' column to numeric values
label_encoder = LabelEncoder()
df['Doctor'] = label_encoder.fit_transform(df['Doctor'])

# Separate X (features) and Y (target)
X = df.drop('Doctor', axis=1)  # All columns except 'Disease'
Y = df['Doctor']  # The 'Disease' column as target

best_accuracy = 0
best_model = None
best_Y_test = None
best_Y_pred = None

# Run the training 10 times
for i in range(10):
    # Split the data into train and test sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=i)

    # Create a Linear Regression model
    lin_reg = LinearRegression()

    # Train the model
    lin_reg.fit(X_train, Y_train)

    # Predict using the test set
    Y_pred = lin_reg.predict(X_test)

    # Round the predictions to nearest integer
    Y_pred_rounded = np.round(Y_pred)

    # Calculate accuracy
    accuracy = accuracy_score(Y_test, Y_pred_rounded)
    print(f"Iteration {i+1}: Model Accuracy = {accuracy * 100:.2f}%")

    # Check if this is the best model
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = lin_reg
        best_Y_test = Y_test
        best_Y_pred = Y_pred_rounded

# Save the best model
with open('best_linear_regression_model.pkl', 'wb') as f:
    pickle.dump(best_model, f)

# Print the test results and actual values of the best model
print("\nBest Model Test Results:")
print(f"Accuracy: {best_accuracy * 100:.2f}%")
print("Actual Values:", best_Y_test.values)
print("Predicted Values:", best_Y_pred)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import pickle

# Load the Excel file
file_path = "Specialist.xlsx" 
df = pd.read_excel(file_path)

# Drop the first row
df = df.drop(index=0)

# Encode the 'Doctor' column to numeric values
label_encoder = LabelEncoder()
df['Doctor'] = label_encoder.fit_transform(df['Doctor'])

# Separate X (features) and Y (target)
X = df.drop('Doctor', axis=1)  # All columns except 'Doctor'
Y = df['Doctor']  # The 'Doctor' column as target

best_accuracy = 0
best_model = None
best_Y_test = None
best_Y_pred = None

# Run the training 10 times
for i in range(10):
    # Split the data into train and test sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=i)

    # Create a Random Forest model
    rf_model = RandomForestClassifier(n_estimators=100, random_state=i)

    # Train the model
    rf_model.fit(X_train, Y_train)

    # Predict using the test set
    Y_pred = rf_model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"Iteration {i+1}: Model Accuracy = {accuracy * 100:.2f}%")

    # Check if this is the best model
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = rf_model
        best_Y_test = Y_test
        best_Y_pred = Y_pred

# Save the best model
with open('best_random_forest_model.pkl', 'wb') as f:
    pickle.dump(best_model, f)

# Convert the numeric labels back to their original text form for better visualization
best_Y_test_text = label_encoder.inverse_transform(best_Y_test)
best_Y_pred_text = label_encoder.inverse_transform(best_Y_pred)

# Print the test results and actual values of the best model
print("\nBest Model Test Results:")
print(f"Accuracy: {best_accuracy * 100:.2f}%")
for actual, predicted in zip(best_Y_test_text, best_Y_pred_text):
    print(f"Actual: {actual}, Predicted: {predicted}")
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download("punkt_tab")
# Define stopwords and unnecessary phrases
stop_words = set(stopwords.words('english'))
unnecessary_phrases = ['i feel', 'i have', 'symptom', 'feeling', 'suffering from']

# Function to match parts of user input to symptoms
def match_symptoms(user_input, symptoms):
    tokens = nltk.word_tokenize(user_input.lower())
    matched_symptoms = []

    # Remove unnecessary phrases
    for phrase in unnecessary_phrases:
        user_input = user_input.replace(phrase, "")
    
    
    for symptom in symptoms:
        for token in tokens:
            if token in symptom.replace("_", " ").lower():
                matched_symptoms.append(symptom)
                break
                
    return matched_symptoms

# User input
user_input = input("Please enter your symptoms: ")

# Match the user input to dataset symptoms
matched_symptoms = match_symptoms(user_input, symptoms)

# Display matched symptoms
if matched_symptoms:
    print(f"Matched Symptoms: {matched_symptoms}")
else:
    print("No symptoms matched.")
```

```python
symptoms = df.columns[:-1].tolist()

# Load the trained model (ensure you saved it earlier)
with open("best_random_forest_model.pkl", "rb") as model_file:
    model = pickle.load(model_file)


# Clean and match symptoms
def clean_input(user_input):
    user_input = user_input.lower()
    
    # Remove unnecessary phrases
    for phrase in unnecessary_phrases:
        user_input = user_input.replace(phrase, "")
    
    # Tokenize and remove stopwords
    tokens = nltk.word_tokenize(user_input)
    cleaned_tokens = [token for token in tokens if token not in stop_words]
    
    return cleaned_tokens

def match_symptoms(user_input, symptoms):
    cleaned_tokens = clean_input(user_input)
    matched_symptoms = []
    
    for symptom in symptoms:
        for token in cleaned_tokens:
            if token in symptom.replace("_", " ").lower():
                matched_symptoms.append(symptom)
                break
                
    return matched_symptoms

# Create input vector for model based on matched symptoms
def create_input_vector(matched_symptoms, symptoms):
    input_vector = [0] * len(symptoms)  # Initialize input vector with zeros
    for symptom in matched_symptoms:
        if symptom in symptoms:
            idx = symptoms.index(symptom)
            input_vector[idx] = 1  # Set 1 for matched symptoms
    return input_vector

# User input
user_input = input("Please enter your symptoms: ")

# Match symptoms and prepare input vector
matched_symptoms = match_symptoms(user_input, symptoms)
print(matched_symptoms)
input_vector = create_input_vector(matched_symptoms, symptoms)

# Convert input_vector to a DataFrame for prediction
input_df = pd.DataFrame([input_vector], columns=symptoms)

# Predict using the trained model
predicted_doctor_num = model.predict(input_df)[0]

print(predicted_doctor_num)
# Convert predicted disease number back to text
predicted_doctor = label_encoder.inverse_transform([predicted_doctor_num])

# Display result
print(f"Predicted Doctor: {predicted_doctor}")
```
