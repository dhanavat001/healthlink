```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pickle

# Load dataset
data = pd.read_csv("healthcare.csv")

# Convert prognosis column to numeric using LabelEncoder
label_encoder = LabelEncoder()
data['prognosis'] = label_encoder.fit_transform(data['prognosis'])

# Split the data into X (features) and Y (target)
X = data.drop('prognosis', axis=1)
Y = data['prognosis']

# Variables to store the best model and highest accuracy
best_accuracy = 0
best_model = None

# Run the training and evaluation process 10 times
for i in range(10):
    # Split the dataset into training and testing sets (90% train, 10% test)
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=i)

    # Initialize the KNN classifier
    knn = KNeighborsClassifier(n_neighbors=5)

    # Train the model
    knn.fit(X_train, Y_train)

    # Predict on the test set
    Y_pred = knn.predict(X_test)

    # Convert numeric values back to original form (text)
    Y_test_original = label_encoder.inverse_transform(Y_test)
    Y_pred_original = label_encoder.inverse_transform(Y_pred)

    # Calculate accuracy
    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"Iteration {i+1}, Accuracy: {accuracy * 100:.2f}%")

    # Check if this is the best model so far
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = knn

        # Save the best model using pickle
        with open("best_knn_model.pkl", "wb") as f:
            pickle.dump(best_model, f)

# Load the best model (for demonstration purposes)
with open("best_knn_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)

# Predict again using the best model
Y_pred_best = loaded_model.predict(X_test)

# Convert numeric values back to text
Y_pred_best_original = label_encoder.inverse_transform(Y_pred_best)
Y_test_best_original = label_encoder.inverse_transform(Y_test)

# Print the actual vs predicted values for the best model
print("\nBest Model: Actual vs Predicted:")
for actual, predicted in zip(Y_test_best_original, Y_pred_best_original):
    print(f"Actual: {actual}, Predicted: {predicted}")

# Print the best accuracy achieved
print(f"\nBest Model Accuracy: {best_accuracy * 100:.2f}%")
```
